```python
import pandas as pd

# Load the raw data
df = pd.read_csv('../data/raw/nhl_game_data.csv')

# Convert scores to integers
df['Visitor Goals'] = df['Visitor Goals'].astype(int)
df['Home Goals'] = df['Home Goals'].astype(int)

# Feature engineering: create goal difference
df['Goal Difference'] = df['Visitor Goals'] - df['Home Goals']

# Save the cleaned data to a CSV file
df.to_csv('../data/processed/cleaned_nhl_game_data.csv', index=False)

# Display the first few rows of the cleaned dataframe
df.head()
```
